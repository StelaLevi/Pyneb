```python
import pyneb as pn  # Importando PyNeb
```

```python
import matplotlib.pyplot as plt
```

```python
# Verificando a versão do PyNeb
print(pn.__version__)
```

```python
# Criando um objeto Atom para o íon O++ ([O III])
O3 = pn.Atom('O', 3)
```

```python
print(O3)
```

```python
# Informações básicas sobre o objeto Atom
print(O3.name)  # Nome do elemento
print(O3.spec)  # Número de ionização
print(O3.Z)     # Número atômico
```

```python
# Mostrando os níveis de energia
print('{} eV - {} eV'.format(O3.IP, O3.IP_up))
print('Energy of the 4th level = {:.2f} eV.'.format(O3.getEnergy(4, unit='eV')))
```

```python
# Listando métodos disponíveis para O3
print(dir(O3))
```

```python
# Obtendo ajuda sobre um método específico
help(O3.getEnergy)
```

```python
# Imprimindo informações sobre o íon
O3.printIonic()
O3.printIonic(tem=10000., den=1e3, printA=True, printPop=True, printCrit=True)
```

```python
# Plotando emissividades
f, ax = plt.subplots(figsize=(8, 6))
O3.plotEmiss(tem_min=1000, tem_max=20000, ionic_abund=1.0, den=1e3, style='-', legend=True, ax=ax)
plt.show()
```

```python
# Plotando diagrama de Grotrian
f, ax = plt.subplots(figsize=(8, 6))
O3.plotGrotrian(tem=1e4, den=1e2, thresh_int=1e-3, unit='eV', ax=ax)
plt.show()
```

```python
# Acessando dados atômicos
print(O3.getA())
print(O3.getOmegaArray().shape)
print(O3.getOmega(tem=1e4))
```

```python
# Obtendo ajuda sobre getOmega
help(O3.getOmega)
```
